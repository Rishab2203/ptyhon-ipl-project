```python
import pandas as pd
matches_df = pd.read_csv("../data/matches.csv")
season_id_df = matches_df[["id","season"]]
deliveries_df = pd.read_csv("../data/deliveries.csv")
deliveries_df.rename(columns={"match_id":"id"},inplace=True)

# Merging the two dataframes(right join) on match_id.

df = season_id_df.merge(deliveries_df,on="id",how="right")
df.head()
```

```python
# filtering the rows by season and getting the desired rows 
df.loc[df["season"]==2016,["bowling_team","extra_runs"]].groupby("bowling_team")["extra_runs"].sum().reset_index()
```
